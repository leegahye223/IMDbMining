# Question 2

```r
edgelist_file <- 'mergeNew.txt'
g <- read.graph(file = edgelist_file, format = "ncol", directed = T)
```

```r
inDeg <- strength(g, mode = "in")
print(max(inDeg))
```

```r
hist(inDeg, breaks = seq(0,721,1),
     main = NA, xlab = "In-degree", ylab= "Count");
```

# Question 4

```r
visit_prob <- page_rank(g, damping = 0.85)$vector
```

```r
sortedProb <- sort(visit_prob, decreasing = T, index.return = T)
ids <- sortedProb$ix[1:10]
```

```r
print(ids)
```

```r
print(sortedProb$x[1:10])
```
