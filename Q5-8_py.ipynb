```python
actorFile  = open('actor_movies.txt','r',encoding = "ISO-8859-1").read().split('\n')
acctressFile = open('actress_movies.txt','r',encoding = "ISO-8859-1").read().split('\n')


# merging movies
movie_dict = {}

for i in range(0, len(actorFile)):
    l =  actorFile[i].split('\t\t')
    if not l[0]:
        continue
    if len(l[1:]) < 10:
        continue
    for movie in l[1:]:
        pos = movie.find(')')
        movie = movie[:pos + 1]
        movie.strip()
        if not movie:
            continue
        actor = l[0].strip()
        if movie in movie_dict:
            movie_dict[movie].append(actor)
        else:
            movie_dict[movie] = [actor]
    
 #=========================================================   

for i in range(0, len(acctressFile)):
    l =  acctressFile[i].split('\t\t')
    if not l[0]:
        continue
    if len(l[1:]) < 10:
        continue
    for movie in l[1:]:
        pos = movie.find(')')
        movie = movie[:pos + 1]
        movie.strip()
        if not movie:
            continue
        actor = l[0].strip()
        if movie in movie_dict:
            movie_dict[movie].append(actor)
        else:
            movie_dict[movie] = [actor]
    
print("total number of movies: ", len(movie_dict))
```

```python
movie_genre_dict = dict()
MovGenre = open("movie_genre.txt", "r", encoding='ISO-8859-1').read().split('\n')
#output = open("movie_genre_new.txt", "w")
genre_name = []

for line in MovGenre:
    line = line.split('\t\t')
    if len(line) < 2:
        continue
    genre_name.append(line[1])   
Unigenre = list(set(genre_name))
gsum = np.zeros(29)

count = 0
for i in range(0, len(MovGenre)):
    l = MovGenre[i].split('\t\t')
    if len(l) > 1:
        if l[0] not in moviemap:
            continue
        movie_genre_dict[moviemap[l[0]]] = Unigenre.index(l[1])
        gsum[Unigenre.index(l[1])] += 1
        count += 1
q = gsum / count
        #output.write(str(l[0])+"\t"+str(l[1])+"\n")
#output.close()
```

```python
output=open("movie_edge_list2.txt", "w")
edge_list = set()

for i in range(len(actor_list)): 
    for actor in actor_list[i]:
        MoviesPlayed = actor_dict[MapActor[actor]]
        for movie in MoviesPlayed:
            if movie not in movie_dict:
                continue
            j = moviemap[movie]
            if j > i:
                AInteSect = set(actor_list[i]).intersection(set(actor_list[j]))
                n = len(list(AInteSect))
                if n == 0:
                    continue                
                
                wij = (n / float(len(actor_list[i]) + len(actor_list[j]) - n))
                edge_list.add((i,j,wij)) 

edge_list = list(edge_list)


for i in range(len(edge_list)):
    output.write(str(edge_list[i][0])+"\t\t"+str(edge_list[i][1])+"\t\t"+str(edge_list[i][2])+"\n")
                
output.close()   
```
